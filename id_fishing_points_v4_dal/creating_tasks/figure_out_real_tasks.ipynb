```python
import json
mypath = "../../data/vessels_20160927/"
import numpy as np
from os import listdir
from os.path import isfile, join
files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
import csv
tasksA = []
tasksB = []
tasksA_mmsi = {}
tasksB_mmsi = {}

with open("../tasks_20160927.csv","rb") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        try:
            j = row['mmsi']+"_"+row['year']+"_"+row['month']+".json"
            f = open(mypath+j, 'rb')
            f = json.loads(f.read())
            sogs = np.array(f['sogs'])
            lons = np.array(f['lons'])
            lats = np.array(f['lats'])
            mmsi = row['mmsi']
            year = row['year']
            month = row['month']
            vesselType = row['vesselType']
    #         print lats.max() - lats.min(), lons.max()-lons.min(),len(np.where( sogs > 4 )[0]) 
            if len(np.where( sogs > 4 )[0]) > 100:
                if mmsi not in tasksA_mmsi:
                    tasksA_mmsi[mmsi] = 0
                if tasksA_mmsi[mmsi] < 2:
                    tasksA.append([mmsi,year,month,vesselType])
                else:
                    tasksB.append([mmsi,year,month,vesselType])

                tasksA_mmsi[mmsi]+=1
        except:
            pass
            
    
#         break
```

```python
with open("../tasks_20160927A.csv","wb") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["mmsi","year","month","vesselType"])
    writer.writerows(tasksA)

with open("../tasks_20160927B.csv","wb") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["mmsi","year","month","vesselType"])
    writer.writerows(tasksB)
```
